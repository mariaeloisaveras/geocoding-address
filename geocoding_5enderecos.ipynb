```python
from geopy.extra.rate_limiter import RateLimiter
import pandas as pd
from geopy.geocoders import Nominatim
```

```python
# http request
locator = Nominatim(user_agent="myGeocoder")
```

```python
# abrir o csv (separador= ;)
df = pd.read_csv('enderecos_teste.csv',sep= ';')
df.head()
```

```python
#  concatenar os endereços
df['endereco'] = df['nome'].astype(str) + ',' + \
                 df['num'].astype(str) + ',' +'São Paulo' + ',' + 'São Paulo' + ',' + 'Brasil'

df.head()
```

```python
# geocodificar os endereços
geocode = RateLimiter(locator.geocode, min_delay_seconds=1)
df['localizacao'] = df['endereco'].apply(geocode)
df['pontos'] = df['localizacao'].apply(lambda loc: tuple(loc.point) if loc else None)
```

```python
df.head()
```

```python
#  criar colunas contendo latitude, longitude e altitude
df[['latitude', 'longitude', 'altitude']] = pd.DataFrame(df['pontos'].tolist(), index=df.index)
df.head()
```

```python
# criar tabela conendo apenas o endereço, a latitude, a longitude e a altitude
df_clean= df.drop(['nome', 'num','endereco', 'pontos'], axis= 1)
df_clean.head()
```

```python
# salvar como .csv 
df.to_csv(r'enderecos_geocoding.csv')
```
